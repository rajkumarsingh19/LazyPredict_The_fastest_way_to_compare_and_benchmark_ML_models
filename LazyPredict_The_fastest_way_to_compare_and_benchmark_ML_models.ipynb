# **Classification**

```python
# Import libraries
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load dataset
data = load_breast_cancer()
X = data.data
y= data.target

# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =42)

# Defines and builds the lazyclassifier
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models_train,predictions_train = clf.fit(X_train, X_train, y_train, y_train)
models_test,predictions_test = clf.fit(X_train, X_test, y_train, y_test)

# Prints the model performance
models_train
```

```python
models_test
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sort models by Accuracy in descending order
models_sorted = models_train.sort_values('Accuracy', ascending=False)

# Create clean, crisp visualization
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Create bar plot with sorted data
ax = sns.barplot(x=models_sorted.index, y="Accuracy", data=models_sorted,
                 palette="viridis", alpha=0.8)

# Customize the plot
plt.xticks(rotation=90, fontsize=10)
plt.title('Model Accuracy Comparison (Sorted)', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Models', fontsize=12, fontweight='bold')
plt.ylabel('Accuracy', fontsize=12, fontweight='bold')

# Add value labels on bars
for i, v in enumerate(models_sorted['Accuracy']):
    ax.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom', fontsize=9, rotation=0)

# Tight layout
plt.tight_layout()
plt.show()

# Print top 10 models
print(" TOP 10 MODELS:")
print("=" * 40)
for i, (model, acc) in enumerate(models_sorted.head(10).iterrows(), 1):
    print(f"{i:2d}. {model:<25} {acc['Accuracy']:.3f}")
```

# **Regression**

```python
# Import libraries
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

# Load dataset
housing = fetch_california_housing()
X = housing.data
y= housing.target

# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =42)

# Defines and builds the lazyclassifier
reg = LazyRegressor(verbose=0,ignore_warnings=False, custom_metric=None)
models_train,predictions_train = reg.fit(X_train, X_train, y_train, y_train)
models_test,predictions_test = reg.fit(X_train, X_test, y_train, y_test)

# Prints the model performance
models_train
```
